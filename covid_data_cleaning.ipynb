### Policy Data Cleaning

```python
covid_df.drop(columns=['Unnamed: 0','Unnamed: 0.1','country_name','alpha_3_code','region_code'],inplace=True)
```

```python
#drop columns with over 10% null
covid_df.drop(columns=['school_closing_flag', 'school_closing_notes',
                     'workplace_closing_flag', 'workplace_closing_notes',
                     'cancel_public_events_flag', 'cancel_public_events_notes',
                     'restrictions_on_gatherings_flag', 'restrictions_on_gatherings_notes',
                     'close_public_transit_flag', 'close_public_transit_notes',
                     'stay_at_home_requirements_flag', 'stay_at_home_requirements_notes',
                     'restrictions_on_internal_movement_flag', 'restrictions_on_internal_movement_notes',
                     'international_travel_controls_notes',
                     'income_support_flag',  'income_support_notes',
                     'debt_contract_relief_notes',
                     'fiscal_measures_notes',
                     'international_support_notes',
                     'public_information_campaigns_flag', 'public_information_campaigns_notes',
                     'testing_policy_notes',
                     'contact_tracing_notes',
                     'emergency_healthcare_investment_notes',
                     'vaccine_investment_notes',
                     'misc_wildcard', 'misc_wildcard_notes'],
            inplace=True)
```

```python
# Fill with average for state
covid_df['school_closing'].fillna(covid_df['school_closing'].mean(), inplace=True)
covid_df['workplace_closing'].fillna(covid_df['workplace_closing'].mean(), inplace=True)
covid_df['cancel_public_events'].fillna(covid_df['cancel_public_events'].mean(), inplace=True)
covid_df['close_public_transit'].fillna(covid_df['close_public_transit'].mean(), inplace=True)
covid_df['stay_at_home_requirements'].fillna(covid_df['stay_at_home_requirements'].mean(), inplace=True)
covid_df['restrictions_on_gatherings'].fillna(covid_df['restrictions_on_gatherings'].mean(), inplace=True)
covid_df['restrictions_on_internal_movement'].fillna(covid_df['restrictions_on_internal_movement'].mean(), inplace=True)
covid_df['international_travel_controls'].fillna(covid_df['international_travel_controls'].mean(), inplace=True)
covid_df['debt_contract_relief'].fillna(covid_df['debt_contract_relief'].mean(), inplace=True)
covid_df['public_information_campaigns'].fillna(covid_df['public_information_campaigns'].mean(), inplace=True)
covid_df['testing_policy'].fillna(covid_df['testing_policy'].mean(), inplace=True)
covid_df['contact_tracing'].fillna(covid_df['contact_tracing'].mean(), inplace=True)
covid_df['stringency_index'].fillna(covid_df['stringency_index'].mean(), inplace=True)
```

```python
# Replace nulls with zeros
covid_df['income_support'].fillna(0, inplace=True)
covid_df['fiscal_measures'].fillna(0, inplace=True)
covid_df['international_support'].fillna(0, inplace=True)
covid_df['vaccine_investment'].fillna(0, inplace=True)
covid_df['emergency_healthcare_investment'].fillna(0, inplace=True)
covid_df['confirmed_cases'].fillna(0, inplace=True)
covid_df['deaths'].fillna(0, inplace=True)
```

### Mobility Data Cleaning

```python
#filling in nulls with median
covid_df['retail_and_recreation_percent_change_from_baseline'].fillna(covid_df['retail_and_recreation_percent_change_from_baseline'].median(), inplace=True)
```

```python
#filling in nulls with median
covid_df['grocery_and_pharmacy_percent_change_from_baseline'].fillna(covid_df['grocery_and_pharmacy_percent_change_from_baseline'].median(), inplace=True)
```

```python
#filling in nulls with median
covid_df['parks_percent_change_from_baseline'].fillna(covid_df['parks_percent_change_from_baseline'].median(), inplace=True)
```

```python
#filling in nulls with median
covid_df['transit_stations_percent_change_from_baseline'].fillna(covid_df['transit_stations_percent_change_from_baseline'].median(), inplace=True)
```

```python
#filling in nulls with median
covid_df['workplaces_percent_change_from_baseline'].fillna(covid_df['transit_stations_percent_change_from_baseline'].median(), inplace=True)
```

```python
#filling in nulls with median
covid_df['residential_percent_change_from_baseline'].fillna(covid_df['residential_percent_change_from_baseline'].median(), inplace=True)
```
